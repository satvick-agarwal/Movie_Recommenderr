```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("../data/TMDB_all_movies.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df["overview"].head()
```

```python
df = df.dropna(subset=["overview","title"])
```

```python
df.shape
```

```python
df = df[
    [
        "id",
        "title",
        "vote_average",
        "overview",
        "popularity",
        "genres",
        "spoken_languages",
        "imdb_rating"
    ]
].copy()
```

```python
df.shape
```

```python
df["genres"].head()
```

```python
df["genres"].iloc[0]
```

```python
df["genres"] = df["genres"].fillna("").str.strip()
```

```python
df["genres"].head(5)
(df["genres"].str.len() > 0).sum()
```

```python
df = df[
    df["spoken_languages"]
    .fillna("")
    .str.contains("english|हिन्दी|hindi", case=False, regex=True)
]
```

```python
df["spoken_languages"].head()
```

```python
df.loc[:, 'imdb_rating'] = df['imdb_rating'].fillna(df['imdb_rating'].mean())
```

```python
df['combined_text'] = (
    df['overview'] + " " + df['genres']
).str.lower()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df=df.to_csv("../data/cleaned_movies.csv", index=False)
```
